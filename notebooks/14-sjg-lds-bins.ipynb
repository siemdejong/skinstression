```python
import sys
sys.path.append("../src")
from ds.dataset import THGStrainStressDataset
import numpy as np
from tqdm.contrib import tenumerate
import plotly.graph_objects as go
import pandas as pd
```

```python
csv_targets = pd.read_csv('../data/z-stacks/logistic_targets.csv')
csv_targets = csv_targets.drop([1, 10])[["a", "k", "xc"]].to_numpy() # drop index 7 and 16 (at row 1 and 10), because their images are too small.
```

```python
param_roi = {
    "a": np.arange(0, 15, 0.01),
    "k": np.arange(0, 50, 0.1),
    "xc": np.arange(1, 1.5, 0.001),
}
param_ks = {
    "a": 30,
    "k": 30,
    "xc": 30,
}
param_sigma = {
    "a": 3,
    "k": 3,
    "xc": 3,
}
```

```python
scaled_weights, num_per_target, eff_num_per_target, edges = THGStrainStressDataset._prepare_weights(csv_targets, "sqrt_inv", True, 'gaussian', param_roi, param_ks, param_sigma, ["a", "k", "xc"])
```

```python
df_a = pd.DataFrame()
df_a['emp'] = num_per_target[0]
df_a['lds'] = eff_num_per_target[0]
df_a['edges'] = edges[0]

fig_a = go.Figure()
fig_a.add_trace(go.Bar(x=df_a['edges'], y=df_a["lds"], name="lds"))
fig_a.add_trace(go.Bar(x=df_a['edges'], y=df_a["emp"], name="emp"))
fig_a.update_layout(barmode='overlay')
fig_a.show()

df_k = pd.DataFrame()
df_k['emp'] = num_per_target[1]
df_k['lds'] = eff_num_per_target[1]
df_k['edges'] = edges[1]

fig_k = go.Figure()
fig_k.add_trace(go.Bar(x=df_k['edges'], y=df_k["lds"], name="lds"))
fig_k.add_trace(go.Bar(x=df_k['edges'], y=df_k["emp"], name="emp"))
fig_k.update_layout(barmode='overlay')
fig_k.show()

df_xc = pd.DataFrame()
df_xc['emp'] = num_per_target[2]
df_xc['lds'] = eff_num_per_target[2]
df_xc['edges'] = edges[2]

fig_xc = go.Figure()
fig_xc.add_trace(go.Bar(x=df_xc['edges'], y=df_xc["lds"], name="lds"))
fig_xc.add_trace(go.Bar(x=df_xc['edges'], y=df_xc["emp"], name="emp"))
fig_xc.update_layout(barmode='overlay')
fig_xc.show()
```
