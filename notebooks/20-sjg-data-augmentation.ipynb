```python
from PIL import Image
from torchvision.transforms import ColorJitter, functional
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "/scistor/guest/sjg203/projects/shg-strain-stress/data/preprocessed/z-stacks/15/19.bmp"
```

```python
img = Image.open(image_path)
fig, ax = plt.subplots(5, 4, figsize=(10, 10))
for i, brightness in enumerate(np.arange(0, 2, 0.1)):
    ax.flat[i].set_title(f"{brightness:.2f}")
    ax.flat[i].imshow(functional.adjust_brightness(img, brightness))

plt.tight_layout()
plt.show()
```
