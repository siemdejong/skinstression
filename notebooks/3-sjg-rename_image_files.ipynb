```python
import os
import numpy as np
```

```python
os.chdir('../data/z-stacks/')
print(os.getcwd())
```

```python
# NOTE: THIS CHANGES MANY FILES IN THE DIRECTORY SPECIFIED ABOVE.
# PROCEED CAREFULLY.
# MAY ALSO SHUFFLE IMAGES.

for root, dirs, files in os.walk(os.getcwd()):
    for new_name, old_name in enumerate(dirs + files):
        if os.path.isdir(old_name):
            continue
        _, f_ext = os.path.splitext(old_name)
        if f_ext not in ["bmp"]:
            continue
        new_name = f'{new_name}{f_ext}'
        os.rename(os.path.join(root, old_name), os.path.join(root, new_name))
```
